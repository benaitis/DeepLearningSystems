# The most popular activation functions

Non-linear activation functions are used to increase model flexibility allowing to solve non-linear problems.

A function is non-linear if the slope (derivative) isn't constant.

More information on activation functions in Keras you can find here:

[https://keras.io/api/layers/activations/](https://keras.io/api/layers/activations/)

## ReLU
Rectified Linear Unit (**ReLU**) - the most popular activation function, used in majority of nets. Default choice for hidden layers. Network learns much faster because gradient stays big. The ReLU function is non-linear around 0, but the slope (derivative) is always either 0 (for negative values) or 1 (for positive values). That's a very limited type of non-linearity.

$f(x) = max(0,x)$

```tf.keras.activations.relu(x, alpha=0.0, max_value=None, threshold=0)```

````python
import numpy as np #library for working with matrices
#%matplotlib notebook
%matplotlib inline
import matplotlib.pyplot as plt #plotting library
````

````python
x = list(np.linspace(-10, 10, num=100))
#print(x)
````

````python
y = [max(0, i) for i in x]

#y = []
#for i in x:
#    y.append(max(0, i))
print(y)
````

````python
plt.figure()
plt.grid()
plt.title("ReLU")
plt.xlabel("x")
plt.ylabel("y")
plt.plot(x, y);
````

## Leaky ReLU
Leaky Rectified Linear Unit (**Leaky ReLU**) - variation of the most popular activation function **ReLU** with slight non-zero slope for negative input, helps training to converge a bit faster. $\alpha$ controls the negative input slope:

$f(x) = \max(\alpha x,x)$

```tf.keras.activations.relu(x, alpha=0.3, max_value=None, threshold=0)```

````python
y = [max(0.3 * i, i) for i in x]
print(y)
````

````python
plt.figure()
plt.grid()
plt.title("Leaky ReLU")
plt.xlabel("x")
plt.ylabel("y")
plt.plot(x, y);
````

## Sigmoid

Another popular function because returns values in range [0;1]. For small values (<-5), sigmoid returns a value close to zero, and for large values (>5) the result of the function gets close to 1.

$f(x) = \frac{1}{1 + e^{-x}}$

```tf.keras.activations.sigmoid(x)```

````python
y = [1/(1 + np.exp(-i)) for i in x]
print(y)
````

````python
plt.figure()
plt.grid()
plt.title("Sigmoid")
plt.xlabel("x")
plt.ylabel("y")
plt.plot(x, y);
````

## Softmax

Popular function in classification problems. Usually used in the last layer. Converts values vector into probabilities vector. Output values are in range [0;1] with sum of them equal to 1.

$f(x) = \frac{e^x}{\sum(e^x)}$


```tf.keras.activations.softmax(x, axis=-1)```

````python
r = [0.2, 0.3, -0.5, -0.01]
exp_r = [np.exp(i) for i in r]
print(exp_r)
sum_exp_r = np.sum(exp_r)
print(sum_exp_r)
prob_r = [i / sum_exp_r for i in exp_r]
print(prob_r)
print(np.sum(prob_r))
````

## Softplus

Similar to ReLU, smooth, requires more resources to calculate.


$f(x) = \log(e^{x} + 1)$

```tf.keras.activations.softplus(x)```

````python
y = [np.log(np.exp(i)+1) for i in x]
````

````python
plt.figure()
plt.grid()
plt.title("Softplus")
plt.xlabel("x")
plt.ylabel("y")
plt.plot(x, y);
````

## Softsign

Similar to tanh with much slower saturation.


$f(x) = \frac{x}{| x | + 1}$

```tf.keras.activations.softsign(x)```

````python
y = [i / (np.abs(i)+1) for i in x]
````

````python
plt.figure()
plt.grid()
plt.title("Softsign")
plt.xlabel("x")
plt.ylabel("y")
plt.plot(x, y);
````

## Tanh

Similar to sigmoid. Output values are centered around 0 and in range (-1;1). Function quickly becomes saturated near input values <-2.5 and >2.5.


$f(x) = \tanh(x)$

```tf.keras.activations.tanh(x)```

````python
y = [np.tanh(i) for i in x]
````

````python
plt.figure()
plt.grid()
plt.title("tanh")
plt.xlabel("x")
plt.ylabel("y")
plt.plot(x, y);
````

## SELU
Scaled Exponential Linear Unit (**SELU**). Modified version of **ELU** with some benefits. Works with weight initializer ```tf.keras.initializers.LecunNormal```. The values of $\alpha$ and $scale$ are chosen so that the mean and variance of the inputs are preserved between two consecutive layers as long as the weights are initialized correctly. Also number of input units should be "large enough".

$
f(x) =
  \begin{cases}
    scale * x       & \quad \text{if } x > 0\\
    scale * \alpha * (e^{x} - 1)  & \quad \text{if } x <= 0
  \end{cases}
$

$\alpha=1.67326324; scale=1.05070098$

```tf.keras.activations.selu(x)```

````python
y = [1.05070098 * i if i > 0 else 1.05070098 * 1.67326324 * (np.exp(i) - 1) for i in x]
````

````python
plt.figure()
plt.grid()
plt.title("SELU")
plt.xlabel("x")
plt.ylabel("y")
plt.plot(x, y);
````

## ELU
Exponential Linear Unit (**ELU**). The value of $\alpha$ controls saturation for negative input.

$
f(x) =
  \begin{cases}
    x       & \quad \text{if } x > 0\\
    \alpha * (e^{x} - 1)  & \quad \text{if } x <= 0
  \end{cases}
$

```tf.keras.activations.elu(x, alpha=1.0)```

````python
y = [i if i > 0 else 1. * (np.exp(i) - 1) for i in x]
````

````python
plt.figure()
plt.grid()
plt.title("ELU")
plt.xlabel("x")
plt.ylabel("y")
plt.plot(x, y);
````

## Exponential

$f(x) = e^{x}$

```tf.keras.activations.exponential(x)```

````python
y = [np.exp(i) for i in x]
````

````python
plt.figure()
plt.grid()
plt.title("Exponential")
plt.xlabel("x")
plt.ylabel("y")
plt.plot(x, y);
````

# Other activation functions

Some activation functions are presented as layers and you can find more information about them here:
[https://keras.io/api/layers/activation_layers/](https://keras.io/api/layers/activation_layers/)

There are also advanced activation functions which can learn their parameters. For example **LeakyReLU** can learn negative input slope $\alpha$. Another function which can learn their parameters is Parametric ReLU (**PReLU**).

# Conclusions

Keras framework provides a lot of already implemented activation functions. Also the framework provides possibility to create custom activations by taking mathematical functions directly from Tensorflow.

In practice however usually two to three activation functions are used. In image classification and object detection problems **ReLU** and **Softmax** are the most popular.
